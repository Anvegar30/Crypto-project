```python
# get all symbol prices
prices_binance = client.get_all_tickers()

# Save in pandas dataframe
binance = pd.DataFrame(prices_binance)

# Rename columns
binance = binance.rename(columns = {'symbol': 'display_name', 'price': 'price'}, inplace = False)
```
